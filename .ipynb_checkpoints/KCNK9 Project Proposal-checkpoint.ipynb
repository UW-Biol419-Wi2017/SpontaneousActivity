```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import csv
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
Data = pd.read_csv('Spont Act Data_Expression_Rows_Combo.csv')
```

```python
data_array = Data.ix[:,5:].values
print(data_array.dtype)
print(data_array)
```

```python
# KMeans Cluster on Expression dataset with 5 clusters
km = KMeans(n_clusters = 5)
km.fit(data_array)

group = km.labels_

plt.scatter(data_array[:,0],data_array[:,10],alpha=0.5, c=['r','b','g','c','y'])
plt.xlim([-0.2,6])
plt.ylim([-0.2,7])
```
