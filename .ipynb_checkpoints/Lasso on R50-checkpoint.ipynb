```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn import linear_model, datasets
from sklearn.linear_model import Lasso
%matplotlib inline
```

```python
data = pd.read_csv('ExpressionR50.csv', header =None)
rows = pd.read_csv('RowsR50.csv')
age = np.array([12, 13, 13, 12, 16, 17, 21, 21, 24, 35, 37])
gene_id = rows.ix[:,2]
data_id = pd.concat([data, gene_id], axis=1)
new_data = data_id[data_id['gene-id']!= 0]
clean_data = new_data.ix[:,:-1]
```

```python
just_data = new_data.ix[:,:11]
genes_pca = PCA()
genes_pc = genes_pca.fit_transform(just_data)
```

```python
print(genes_pca.explained_variance_ratio_)
```

```python
genes_red = genes_pca.transform(just_data)

plt.scatter(genes_red[:,0].flatten(),
            genes_red[:,1].flatten())

plt.xlabel('PC 1')
plt.ylabel('Age')
plt.title('Age vs PC 1')


clf = linear_model.Lasso(alpha=0.1)
clf.fit(genes_red[:,0].flatten().reshape(-1,1), 
        genes_red[:,1].flatten().reshape(-1,1))

print('Lasso Regression:', clf.coef_)
```

```python
clf = linear_model.Lasso(alpha=0.1)
clf.fit(genes_red[:,0].reshape(-1,1), age)

slope = clf.coef_
b = clf.intercept_
fig, ax = plt.subplots()

x_min, x_max = ax.get_xlim()
y_min, y_max = b, b + slope*(x_max-x_min)
ax.plot([x_min, x_max], [y_min, y_max])
ax.set_xlim([x_min, x_max])

print('Lasso Regression:', clf.coef_)
print('Lasso Intercept:', clf.intercept_)
```
