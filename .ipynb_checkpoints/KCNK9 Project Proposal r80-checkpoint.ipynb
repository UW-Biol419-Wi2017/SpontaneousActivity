```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn import svm
from sklearn import datasets
```

```python
Data = pd.read_csv('Spont Act Data_Expression_Rows_Combo.csv')
```

```python
# Creating array with just gene classifer numbers
data_array = Data.ix[:,5:].values
```

```python
# KMeans Cluster on Expression dataset with 5 clusters
km = KMeans(n_clusters = 5)
km.fit(data_array)

group = km.labels_

plt.scatter(data_array[:,0],data_array[:,1],
            alpha=0.8, c=group, cmap='jet')
plt.xlim([0,6])
plt.ylim([0,7])
plt.title('Fig 1: KMeans Clustering')
plt.xlabel('12 PCW')
```

```python
# PCA Expression dataset
pca = PCA()
plot_col = pca.fit_transform(data_array)

plt.scatter(x=plot_col[:,0],y=plot_col[:,1],c=group,cmap='jet')
plt.title('Fig 2: PCA')
```

```python
# Eignvalue Analysis on PCA plot
Cov = np.dot(plot_col.T, plot_col) / len(plot_col)
T,D = np.linalg.eig(Cov)

pc1 = D[:, 1] * T[1]
plt.plot([0, pc1[0]], [0, pc1[1]], 'r-', linewidth=3)

plt.scatter(x=plot_col[:,0],y=plot_col[:,1],c=group,cmap='jet')
plt.title('PCA & Eigenvalue Analysis')
```

```python
# Getting dataframe with just gene IDs
gene_ind = Data.ix[:,1:2]
```

```python
# Creating array of KMeans Predicted Cluster Fit
# Making an array of gene IDs and predicted cluster
Predict = km.fit_predict(data_array)
Data_Predict = np.column_stack((Predict,gene_ind))
```

```python
# Making array into Dataframe for manipulation ease
Gene_Pred_DF = pd.DataFrame(Data_Predict[:,:])
```

```python
# Separating data into clusters
Cluster0 = Gene_Pred_DF[Gene_Pred_DF[0]==0]
Cluster1 = Gene_Pred_DF[Gene_Pred_DF[0]==1]
Cluster2 = Gene_Pred_DF[Gene_Pred_DF[0]==2]
Cluster3 = Gene_Pred_DF[Gene_Pred_DF[0]==3]
Cluster4 = Gene_Pred_DF[Gene_Pred_DF[0]==4]
```

```python
# Getting lists of just gene names in each cluster
Cluster0_genes = Cluster0.iloc[0:,1:]
Cluster1_genes = Cluster1.iloc[0:,1:]
Cluster2_genes = Cluster2.iloc[0:,1:]
Cluster3_genes = Cluster3.iloc[0:,1:]
Cluster4_genes = Cluster4.iloc[0:,1:]
```

```python
# Formating back into array for ease
Zero = Cluster0_genes.as_matrix()
One = Cluster1_genes.as_matrix()
Two = Cluster2_genes.as_matrix()
Three = Cluster3_genes.as_matrix()
Four = Cluster4_genes.as_matrix()
```
